```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = [1,2,3,4,5]
var_1 = [10,20,12,25,20]

df = pd.DataFrame({'var':var,'var_1':var_1})

sns.lineplot(x='var',y='var_1',data=df)
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('penguins.csv').head(20)
```

```python
import matplotlib.pyplot as plt
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=data, hue='sex',style='sex',
             palette='Accent',markers=['o','>'],dashes=False,)
plt.grid()
plt.title('Male vs Female')
plt.show()
```
