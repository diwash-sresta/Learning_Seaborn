```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')
sns.boxplot(df)
plt.show()
```

```python
sns.boxplot(x='day',y ='total_bill',data=df,palette='light:green',
            showmeans=True,meanprops={'markeredgecolor':'w'})
plt.show()
```
