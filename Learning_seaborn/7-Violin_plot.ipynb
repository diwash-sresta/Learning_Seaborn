```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')

sns.violinplot(x= 'day',y='total_bill',data= df)
plt.show()
```

```python
sns.violinplot(x= 'day',y='total_bill',data= df, hue = 'time')
plt.show()
```

```python
sns.violinplot(x= 'day',y='total_bill',data= df, hue = 'time', linewidth=2, palette='Dark2')
plt.show()
```

```python
sns.violinplot(x= 'time',y='total_bill',data= df, order=['Dinner','Lunch'])
plt.show()
```

```python
sns.violinplot(x= 'day',y='total_bill',data= df, hue='sex',split=True)
plt.show()
```

```python
sns.violinplot(x= 'time',y='total_bill',data= df, order=['Dinner','Lunch'],inner='stick')
plt.show()
```

```python
sns.violinplot(x=df['total_bill'])
plt.show()
```
