```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')

fg = sns.FacetGrid(df,col='day',hue='sex')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(df,col='day',hue='sex')
fg.map(plt.bar,'total_bill','tip').add_legend()
plt.show()
```
