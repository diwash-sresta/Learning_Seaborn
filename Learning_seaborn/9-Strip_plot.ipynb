```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')
sns.stripplot(df)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=df,hue='sex',palette='Dark2',edgecolor='w',linewidth=0.5)
plt.show()
```

```python
sns.stripplot(y=df['total_bill'],data=df,hue='sex',jitter=0.5,edgecolor='w',linewidth=0.5)
plt.show()
```
