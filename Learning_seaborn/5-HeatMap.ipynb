```python
df = np.linspace(1,10,10).reshape(2,5)
df
```

```python
sns.heatmap(df,vmin=0,vmax=10,cmap='PuOr',annot=True)
```

```python
arr = np.array([['a0','a1','a2','a3','a4'],['b0','b1','b2','b3','b4']])
arr
```

```python
sns.heatmap(df,vmin=0,vmax=10,cmap='PuOr',annot=arr,fmt='s')
```

```python
y={'fontsize':20,'color':'lime'}
v=sns.heatmap(df,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=10,linecolor='yellow',cbar=False
            ,xticklabels=False,yticklabels=False)
v.set(xlabel='Python',ylabel='Django')
sns.set(font_scale=5)
plt.show()
```
