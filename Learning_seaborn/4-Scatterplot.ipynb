```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset('Penguins').head(20)
df
```

```python
sns.scatterplot(x='bill_length_mm', y = 'bill_depth_mm', data=df, hue='sex')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm', y = 'bill_depth_mm', data=df, hue='sex', style='sex',size='sex',sizes=(20,30), palette='gist_ncar')
plt.show()
```

```python
m={'MALE':'*','FEMALE':'o'}
sns.scatterplot(x='bill_length_mm', y = 'bill_depth_mm', data=df, hue='sex', style='sex',size='sex',sizes=(20,30), palette='gist_ncar', markers=m)
plt.show()
```
